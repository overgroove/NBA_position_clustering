```python
clus1 = []
clus2 = []
clus3 = []
x = float('nan')

for j in range(lineup.shape[0]):
    name = (lineup["player1"][j]+"/"+lineup["Season"][j])
    if name in dict.keys():
        #print(name)
        #print(dict[name])
        clus1.append(dict[name])
    else : 
        clus1.append(x)
        
        
for j in range(lineup.shape[0]):
    name = (lineup["player2"][j][1:]+"/"+lineup["Season"][j])
    if name in dict.keys():
        #print(name)
        #print(dict[name])
        clus2.append(dict[name])
    else : 
        clus2.append(x)
        
for j in range(lineup.shape[0]):
    name = (lineup["player3"][j][1:]+"/"+lineup["Season"][j])
    if name in dict.keys():
        #print(name)
        #print(dict[name])
        clus3.append(dict[name])
    else : 
        clus3.append(x)
```
